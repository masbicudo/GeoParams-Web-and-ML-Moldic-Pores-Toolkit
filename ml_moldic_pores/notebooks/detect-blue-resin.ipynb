```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import numba
```

```python
# Image path
path = f"../data/thin_sections/2-SMC-1-AL_72,99m_n=_2,5x_cesc.jpg"

# Reading an image in default mode:
inputImage = cv.imread(path)
plt.imshow(inputImage[:,:,::-1])

# BGR to CMKY:
inputImageCMYK = colorspace.bgr2cmyk(inputImage)
```

```python
(C, M, Y, K) = (inputImageCMYK[..., 0],
                inputImageCMYK[..., 1],
                inputImageCMYK[..., 2],
                inputImageCMYK[..., 3])
```

```python
cv.imwrite("../out/c.jpg", C)
cv.imwrite("../out/m.jpg", M)
cv.imwrite("../out/y.jpg", Y)
cv.imwrite("../out/k.jpg", K)
```

```python
binaryImage = cv.inRange(
    inputImageCMYK,
    (128,   0,   0,   0),
    (255, 255, 255, 196))
binaryImage
```

```python
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
binaryImage = cv.morphologyEx(binaryImage, cv.MORPH_DILATE, kernel, iterations=1)
binaryImage = cv.morphologyEx(binaryImage, cv.MORPH_ERODE, kernel, iterations=1)
binaryImage = cv.morphologyEx(binaryImage, cv.MORPH_ERODE, kernel, iterations=1)
binaryImage = cv.morphologyEx(binaryImage, cv.MORPH_DILATE, kernel, iterations=1)

plt.imshow(binaryImage, cmap='gray')
cv.imwrite("../out/some.jpg", binaryImage)
porosidade = np.sum(binaryImage/255)/binaryImage.size
print(f"porosidade = {porosidade}")
```
