```python
def adjust_gamma(image, gamma=1.0):
   invGamma = 1.0 / gamma
   table = np.array([
      ((i / 255.0) ** invGamma) * 255
      for i in np.arange(0, 256)])
   return cv2.LUT(image.astype(np.uint8), table.astype(np.uint8))
```

```python
darkenedImage = adjust_gamma(inputImage, 0.5)
plt.imshow(darkenedImage[:,:,::-1])
```

```python
path = f"../out/classificada_01/{image_name}_25_gamma=0.5.jpg"
cv2.imwrite(
    path,
    darkenedImage,
    [cv2.IMWRITE_JPEG_QUALITY, 99]
    )
```
